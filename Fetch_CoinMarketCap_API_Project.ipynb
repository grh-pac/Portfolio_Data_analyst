```python
def api_runner():
    """
    This function makes an API request to fetch cryptocurrency data from CoinMarketCap.
    Note: This example uses Python 2.7 and the python-request library.
    """

    global df  # Global variable to store fetched data

    # Import necessary libraries
    from requests import Request, Session
    from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
    import json
    import pandas as pd

    # Define the URL and parameters for the API request
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'5000',
      'convert':'USD'
    }

    # Define the headers required for the API request
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': 'fc2b2183-8c89-4965-8036-565e620ec2ac',  # Provide your own API key here
    }

    # Create a session and update headers
    session = Session()
    session.headers.update(headers)

    # Try to make the API request and handle exceptions
    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
#       print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)

    # Set pandas option to display all columns
    pd.set_option('display.max_columns', None)
    
    # Normalize the JSON data and append to existing dataframe
    df2 = pd.json_normalize(data['data'])
    df2['timestamp'] = pd.Timestamp.now()  # Add timestamp to the data
    df = df.append(df2)  # Append to global dataframe 'df'
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
import os
from time import time
from time import sleep

# Loop to call api_runner function 333 times
for i in range(333):
    api_runner()  # Call the api_runner function
    print('API Runner completed')  # Print a message indicating completion
    sleep(86400)  # Pause execution for 86400 seconds before the next iteration

exit()  # Exit the script
```
