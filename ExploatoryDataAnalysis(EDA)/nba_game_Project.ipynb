### In This project We analyze NBA game data from the 2010 and 2014 seasons. It includes data preprocessing, visualization, and statistical analysis using Python libraries such as Pandas, NumPy, Matplotlib, and Seaborn. Visualizations include histograms, boxplots, and heatmaps to explore point distributions, team performance, and game location results across seasons.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the NBA games dataset
nba_games = pd.read_csv('nba_games.csv')
```

```python
# Subset the data for the 2010 and 2014 seasons
nba_2010_season = nba_games[nba_games['year_id'] == 2010]
nba_2014_season = nba_games[nba_games['year_id'] == 2014]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot for 2010 Season
axes[0].hist(nba_2010_season[nba_2010_season['fran_id'] == 'Knicks']['pts'], alpha=0.8, density=True, label='Knicks')
axes[0].hist(nba_2010_season[nba_2010_season['fran_id'] == 'Nets']['pts'], alpha=0.8, density=True, label='Nets')
axes[0].legend()
axes[0].set_title("Points Distribution - 2010 Season")
axes[0].set_xlabel("Points Scored")
axes[0].set_ylabel("Density")

# Plot for 2014 Season
axes[1].hist(nba_2014_season[nba_2014_season['fran_id'] == 'Knicks']['pts'], alpha=0.8, density=True, label='Knicks')
axes[1].hist(nba_2014_season[nba_2014_season['fran_id'] == 'Nets']['pts'], alpha=0.8, density=True, label='Nets')
axes[1].legend()
axes[1].set_title("Points Distribution - 2014 Season")
axes[1].set_xlabel("Points Scored")
axes[1].set_ylabel("Density")

plt.show()
```

```python
# Visualize the points scored by teams using a boxplot for the 2010 and 2014 seasons
plt.figure(figsize=(10, 6))
sns.boxplot(data=pd.concat([nba_2010_season, nba_2014_season]), x='fran_id', y='pts', hue='year_id')
plt.title("Points Scored by Teams - 2010 vs 2014 Seasons")
plt.xlabel("Team")
plt.ylabel("Points Scored")
plt.legend(title="Season")
plt.show()
```

```python
# Calculate and visualize the location result frequency for the 2010 and 2014 seasons
location_result_freq_2010 = pd.crosstab(nba_2010_season['game_result'], nba_2010_season['game_location'])
location_result_freq_2014 = pd.crosstab(nba_2014_season['game_result'], nba_2014_season['game_location'])
```

```python
# Subplot 1: Heatmap for 2010 Season Location Result Frequency
plt.subplot(1, 2, 1)  # Create a subplot with 1 row, 2 columns, and index 1
sns.heatmap(location_result_freq_2010, annot=True, fmt="d", cmap="YlGnBu")  # Plot heatmap with annotations, integer formatting, and color map
plt.title("Location Result Frequency - 2010 Season")  # Set the title for the subplot
plt.xlabel("Game Location")  # Set the label for the x-axis
plt.ylabel("Game Result")  # Set the label for the y-axis

# Subplot 2: Heatmap for 2014 Season Location Result Frequency
plt.subplot(1, 2, 2)  # Create a subplot with 1 row, 2 columns, and index 2
sns.heatmap(location_result_freq_2014, annot=True, fmt="d", cmap="YlGnBu")  # Plot heatmap with annotations, integer formatting, and color map
plt.title("Location Result Frequency - 2014 Season")  # Set the title for the subplot
plt.xlabel("Game Location")  # Set the label for the x-axis
plt.ylabel("Game Result")  # Set the label for the y-axis

plt.tight_layout()  # Adjust subplot parameters to provide padding between plots
plt.show()  # Display the plot
```
